```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_ret = pd.read_csv('Return_Data.csv')
```

```python
df_ret = df_ret.drop('Unnamed: 0', axis=1)
```

```python
df_ret.head()
```

```python
df_ret['Date']=pd.to_datetime(df_ret['Date'])
```

```python
df_ret.dtypes
```

```python
start_date = '2009-01-01'
end_date = '2021-12-31'
mask = df_ret['Date'].between(start_date, end_date)
```

```python
df_RET = df_ret[mask]
```

```python
df_RET
```

```python
df_return = pd.melt(df_RET, id_vars='Date', var_name='RIC', value_name='returns_monthly')
```

```python
df_return
```

```python
df_summed = df_return.groupby([df_return['RIC'], df_return['Date'].dt.year])['returns_monthly'].sum()
```

```python
df_RETURN = pd.DataFrame(df_summed).reset_index()
```

```python
df_RETURN
```

```python
df_RETURN.to_csv('yearly_return.csv')
```
