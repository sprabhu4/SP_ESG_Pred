```python
import pandas as pd


combined_data = pd.read_csv('preprocessed data.csv')
```

```python
import plotly.express as px




box1 = px.box(combined_data, y="Total Current Liabilities", width= 500, height= 500, points= 'outliers')
box1.show()

box2 = px.box(combined_data, y="Total Debt", width= 500, height= 500, points= 'outliers')
box2.show()
```

```python
# treatment


# Calculate the IQR
Q1 = combined_data.quantile(0.25)
Q3 = combined_data.quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify the outliers
outliers = combined_data[(combined_data < lower_bound) | (combined_data > upper_bound)].dropna()

# Remove the outliers
df_clean = combined_data[~((combined_data < lower_bound) | (combined_data > upper_bound)).any(axis=1)]
```

```python
# post outlier removal visualization


box3 = px.box(df_clean, y="Total Current Liabilities", width= 500, height= 500, points= 'outliers')
box3.show()

box4 = px.box(df_clean, y="Total Debt", width= 500, height= 500, points= 'outliers')
box4.show()
```

```python
# export data after outlier removal

df_clean.to_csv('clean data.csv')
```
